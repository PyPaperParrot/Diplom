```python
import numpy as np

import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.mixture import GaussianMixture
```

```python
def log_return(data):
    """
    Transform exchange rate's values into log-returns 
    """
    return pd.Series(np.log(data[1:].reset_index(drop=True) / data[:-1].reset_index(drop=True)))

def write_to_hdf5(file_path, file_name, data):
    with h5py.File(f'{file_path + file_name}.hdf5', 'w') as f:
        f.create_dataset('log_return', data=data)
        
def write_to_csv(file_path, file_name, data):
    data.to_csv(f'{file_path + file_name}.csv', index=False)
```

```python
day_data = pd.read_csv('DATA/EURUSD_Candlestick_1_Day_BID_05.05.2003-29.04.2020.csv', usecols=['Local time', 'Close'])
hour_data = pd.read_csv('DATA/EURUSD_Candlestick_1_hour_BID_05.05.2003-29.04.2020.csv', usecols=['Local time', 'Close'])
minute_data = pd.read_csv('DATA/EURUSD_Candlestick_1_minute_BID_29.04.2019-29.04.2020.csv', usecols=['Local time', 'Close'])
```

```python
day_data.insert(loc=0, column="Time", value=pd.to_datetime(day_data['Local time'], utc=True))
hour_data.insert(loc=0, column="Time", value=pd.to_datetime(hour_data['Local time'], utc=True))
minute_data.insert(loc=0, column="Time", value=pd.to_datetime(minute_data['Local time'], utc=True))
```

```python
day_data.drop(['Local time'], axis=1, inplace=True)
hour_data.drop(['Local time'], axis=1, inplace=True)
minute_data.drop(['Local time'], axis=1, inplace=True)
```

# Очистим данные от неторговых дней
В неторговые дни цены на валюту не меняются, оставаясь постоянными в соответствии с ценой закрытия последнего торгового дня / часа / минуты

```python
day_data = day_data[day_data['Time'].dt.dayofweek < 5].reset_index(drop=True)
hour_data = hour_data[hour_data['Time'].dt.dayofweek < 5].reset_index(drop=True)
minute_data = minute_data[minute_data['Time'].dt.dayofweek < 5].reset_index(drop=True)
```

```python
day_data.dtypes
```

```python
day_data['Time'] = pd.to_datetime(day_data.Time)
```

```python
print(f'day data shape: {day_data.shape}')
print(f'hour data shape: {hour_data.shape}')
print(f'minute data shape: {minute_data.shape}')
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Time', y='Close', data=day_data, lw=0.5)
plt.xlabel('Time', size=15)
plt.ylabel('CLose ($)', size=15)
plt.title('Exchange rate EUR/USD', size=15)
plt.savefig(fname='Pictures/Exchange rate EURUSD.png')
plt.show()
```

# Логарифмическая доходность(Log return)
Вычислим доходность по дням / часам / минутам

```python
day_return = log_return(day_data.iloc[:, 1])
hour_return = log_return(hour_data.iloc[:, 1])
minute_return = log_return(minute_data.iloc[:, 1])
```

```python
Day_Returns = pd.DataFrame(day_return)
```

```python
Day_Returns['Time'] = day_data.iloc[:, 0]
Day_Returns.rename(columns={'Close': 'Day return'}, inplace=True)
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Time', y='Day return', data=Day_Returns, lw=0.5)
plt.xlabel('Time', size=15)
plt.ylabel('Day return', size=15)
plt.title('Exchange rate EUR/USD, dayly returns', size=15)
plt.savefig(fname='Pictures/Exchange rate EURUSD_dayly_returns.png')
plt.show()
```

# Запись в файл
Далее, в дипломе, главным объектом исследований станет именно распределение жоходностей по дням / часам / минутам.
Наибольший интерес вызывает исследование хвостов данных распределений, т.к. они имеют далекую от "нормальных" форму.
Мы применим несколько различных подходов для приближения эмпирирческих распределений. В частности, для описания с вытянутой вершиной и "тяжелыми" хвостами хорошо подходят **обобщенное гиперболическое распределение**, **смесь гауссовых** и **распределение Лапласа** 

```python
write_to_csv('DATA/', 'EURUSD_Day_RETURN_05.05.2003-29.04.2020', day_return)
write_to_csv('DATA/', 'EURUSD_Hour_RETURN_05.05.2003-29.04.2020', hour_return)
write_to_csv('DATA/', 'EURUSD_Minute_RETURN_29.04.2019-29.04.2020', minute_return)
```
